```python
# the english list has some weird lemma-*s, so I'll remove them
import pandas as pd
```

```python
with open("lemmatizationTagged/en_lemPOS.csv") as f:
    df = pd.read_csv(f, index_col=0)
```

```python
df.head()
```

```python
ndf = df[df.lemma != "*"]
```

```python
len(df), len(ndf)
```

```python
sdf = df[df.lemma == "*"]
```

```python
print(len(sdf))
sdf.head()
```

```python
lm = pd.merge(ndf, sdf, on="original_form", how="left")
print(len(lm))
lm.head()
```

```python
rm = pd.merge(ndf, sdf, on="original_form", how="right")
print(len(rm))
rm.head()
```

```python
im = pd.merge(ndf, sdf, on="original_form", how="inner")
print(len(im))
im.head()
```

```python
im = pd.merge(ndf, sdf, on="original_form", how="outer")
print(len(im))
im.head()
```
