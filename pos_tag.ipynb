```python
import nltk
import pandas as pd
```

```python
# setting the variables
language = "en"
filename = "{}_top_10".format(language)
```

```python
# gathering the data
df = pd.read_csv("data/lemmatize/{0}/{1}.csv".format(language, filename))
```

```python
# removing the relative position column
df = df.drop("Unnamed: 0", axis=1)
```

```python
#df.head()
```

```python
# creating a list to hold the POS tagged words
# TODO: improve code structure
pos_col = list()

for row in df.iterrows():
    # TODO: fix this workaround of needing a list to POS tag
    word = [row[1][0]]
    #print("word", word)
    if type(word[0]) == type("meString"):
        # fetching the tuple from the list
        tagged_word = nltk.pos_tag(word)[0]
    #print("tag", tagged_word)
    # add to list
        pos_col.append(tagged_word)
    else:
        pos_col.append([word, None])
```

```python
#pos_col
```

```python
# create a Series object
# TODO: this should be able to happen right away
pos = pd.Series(pos_col)
#pos
```

```python
# add the new info to a copy of the dataframe
pdf = pd.concat([df, pos], axis=1, names=["word", "frequency", "POS"])
# rename for prettiness
pdf.columns = ["word", "frequency", "POS"]
```
